```python
## Librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn

import warnings
warnings.filterwarnings('ignore')
from math import nan
```

```python
## Carga de la base de datos

CO = pd.read_csv("CO.csv")
NO = pd.read_csv("NO.csv")
NO2 = pd.read_csv("NO2.csv")
NOx = pd.read_csv("NOx.csv")
O3 = pd.read_csv("O3.csv")
PM25 = pd.read_csv("PM25.csv")
PM10 = pd.read_csv("PM10.csv")
PRS = pd.read_csv("PRS.csv")
RAINF = pd.read_csv("RAINF.csv")
RH = pd.read_csv("RH.csv")
SO2 = pd.read_csv("SO2.csv")
SR = pd.read_csv("SR.csv")
TOUT = pd.read_csv("TOUT.csv")
WD = pd.read_csv("WD.csv")
WS = pd.read_csv("WS.csv")
```

```python
## Se eliminan las columnas 
columnasValidas = ['parametro', 'Fecha','CE', 'CE b']

CO_2 = CO[columnasValidas].copy()
NO_2 = NO[columnasValidas].copy()
NO2_2 = NO2[columnasValidas].copy()
NOx_2 = NOx[columnasValidas].copy()
O3_2 = O3[columnasValidas].copy()
PM25_2 = PM25[columnasValidas].copy()
PM10_2 = PM10[columnasValidas].copy()
PRS_2 = PRS[columnasValidas].copy()
RAINF_2 = RAINF[columnasValidas].copy()
RH_2 = RH[columnasValidas].copy()
SO2_2 = SO2[columnasValidas].copy()
SR_2 = SR[columnasValidas].copy()
TOUT_2 = TOUT[columnasValidas].copy()
WD_2 = WD[columnasValidas].copy()
WS_2 = WS[columnasValidas].copy()
```
