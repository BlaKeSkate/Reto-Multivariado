```python
## Se eliminan las columnas 
columnasValidas = ['parametro', 'Fecha','CE', 'CE b']

CO_2 = CO[columnasValidas].copy()
NO_2 = NO[columnasValidas].copy()
NO2_2 = NO2[columnasValidas].copy()
NOx_2 = NOx[columnasValidas].copy()
O3_2 = O3[columnasValidas].copy()
PM25_2 = PM25[columnasValidas].copy()
PM10_2 = PM10[columnasValidas].copy()
PRS_2 = PRS[columnasValidas].copy()
RAINF_2 = RAINF[columnasValidas].copy()
RH_2 = RH[columnasValidas].copy()
SO2_2 = SO2[columnasValidas].copy()
SR_2 = SR[columnasValidas].copy()
TOUT_2 = TOUT[columnasValidas].copy()
WD_2 = WD[columnasValidas].copy()
WS_2 = WS[columnasValidas].copy()
```

```python
invalidos = ['p','c','d','b','m','l','z','o','s','r','e','a','s','f','h']
contaminantes2 = [CO_2,NO_2,NO2_2,NOx_2 ,O3_2 ,PM25_2,PM10_2 ,PRS_2,RAINF_2,RH_2,SO2_2,SR_2 ,TOUT_2,WD_2,WS_2]
```

```python
def eliminar(data2):
    for i in invalidos:
        indexNames = data2[data2['CE b'] == i].index
        data2.drop(indexNames , inplace=True)
        
for j in contaminantes2: 
    eliminar(j)
```

```python
CO_2 = CO_2.drop(columns=['CE b'])
NO_2 = NO_2.drop(columns=['CE b'])
NO2_2 = NO2_2.drop(columns=['CE b'])
NOx_2 = NOx_2.drop(columns=['CE b'])
O3_2 = O3_2.drop(columns=['CE b'])
PM25_2 = PM25_2.drop(columns=['CE b'])
PM10_2 = PM10_2.drop(columns=['CE b'])
PRS_2 = PRS_2.drop(columns=['CE b'])
RAINF_2 = RAINF_2.drop(columns=['CE b'])
RH_2 = RH_2.drop(columns=['CE b'])
SO2_2 = SO2_2.drop(columns=['CE b'])
SR_2 = SR_2.drop(columns=['CE b'])
TOUT_2 = TOUT_2.drop(columns=['CE b'])
WD_2 = WD_2.drop(columns=['CE b'])
WS_2 = WS_2.drop(columns=['CE b'])
```
