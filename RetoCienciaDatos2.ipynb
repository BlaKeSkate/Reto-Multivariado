```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn


import warnings
warnings.filterwarnings('ignore')
from math import nan
```

```python
CO = pd.read_csv("CO.csv")
NO = pd.read_csv("NO.csv")
NO2 = pd.read_csv("NO2.csv")
NOx = pd.read_csv("NOx.csv")
O3 = pd.read_csv("O3.csv")
PM25 = pd.read_csv("PM25.csv")
PM10 = pd.read_csv("PM10.csv")
PRS = pd.read_csv("PRS.csv")
RAINF = pd.read_csv("RAINF.csv")
RH = pd.read_csv("RH.csv")
SO2 = pd.read_csv("SO2.csv")
SR = pd.read_csv("SR.csv")
TOUT = pd.read_csv("TOUT.csv")
WD = pd.read_csv("WD.csv")
WS = pd.read_csv("WS.csv")
```

```python
## Se eliminan las columnas 
columnasValidas = ['parametro', 'Fecha','CE', 'CE b']

CO_2 = CO[columnasValidas].copy()
NO_2 = NO[columnasValidas].copy()
NO2_2 = NO2[columnasValidas].copy()
NOx_2 = NOx[columnasValidas].copy()
O3_2 = O3[columnasValidas].copy()
PM25_2 = PM25[columnasValidas].copy()
PM10_2 = PM10[columnasValidas].copy()
PRS_2 = PRS[columnasValidas].copy()
RAINF_2 = RAINF[columnasValidas].copy()
RH_2 = RH[columnasValidas].copy()
SO2_2 = SO2[columnasValidas].copy()
SR_2 = SR[columnasValidas].copy()
TOUT_2 = TOUT[columnasValidas].copy()
WD_2 = WD[columnasValidas].copy()
WS_2 = WS[columnasValidas].copy()
```

```python
CO2.head()
```

```python
invalidos = ['p','c','d','b','m','l','z','o','s','r','e','a','s','f','h']
contaminantes2 = [CO_2,NO_2,NO2_2,NOx_2 ,O3_2 ,PM25_2,PM10_2 ,PRS_2,RAINF_2,RH_2,SO2_2,SR_2 ,TOUT_2,WD_2,WS_2]
```

```python
def eliminar(data2):
    for i in invalidos: 
      indexNames = data2[data2['CE b'] == i].index
      data2.drop(indexNames , inplace=True)
        
for j in contaminantes2: 
    eliminar(j)
```

```python
CO_2.head()
```

```python
CO_2 = CO_2.drop(columns=['CE b'])
NO_2 = NO_2.drop(columns=['CE b'])
NO2_2 = NO2_2.drop(columns=['CE b'])
NOx_2 = NOx_2.drop(columns=['CE b'])
O3_2 = O3_2.drop(columns=['CE b'])
PM25_2 = PM25_2.drop(columns=['CE b'])
PM10_2 = PM10_2.drop(columns=['CE b'])
PRS_2 = PRS_2.drop(columns=['CE b'])
RAINF_2 = RAINF_2.drop(columns=['CE b'])
RH_2 = RH_2.drop(columns=['CE b'])
SO2_2 = SO2_2.drop(columns=['CE b'])
SR_2 = SR_2.drop(columns=['CE b'])
TOUT_2 = TOUT_2.drop(columns=['CE b'])
WD_2 = WD_2.drop(columns=['CE b'])
WS_2 = WS_2.drop(columns=['CE b'])
```

```python
df = pd.merge(CO_2, NO_2, on='Fecha', how='outer')
df = pd.merge(df, NO2_2, on='Fecha', how='outer')
df = pd.merge(df, NOx_2, on='Fecha', how='outer')
df = pd.merge(df, O3_2, on='Fecha', how='outer')
df = pd.merge(df, PM25_2, on='Fecha', how='outer')
df = pd.merge(df, PM10_2, on='Fecha', how='outer')
df = pd.merge(df, PRS_2, on='Fecha', how='outer')
df = pd.merge(df, RAINF_2, on='Fecha', how='outer')
df = pd.merge(df, RH_2, on='Fecha', how='outer')
df = pd.merge(df, SO2_2, on='Fecha', how='outer')
df = pd.merge(df, SR_2, on='Fecha', how='outer')
df = pd.merge(df, TOUT_2, on='Fecha', how='outer')
df = pd.merge(df, WD_2, on='Fecha', how='outer')
df = pd.merge(df, WS_2, on='Fecha', how='outer')

df.head()
```

```python
df = df.drop(columns=['parametro_x'])
df = df.drop(columns=['parametro_y'])
df = df.drop(columns=['parametro'])
```
